```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datas = pd.read_csv("data.csv")
datas.head(5)
```

```python
X =  datas.iloc[:, 1:2].values
y = datas.iloc[:,2].values
```

```python
X = datas.iloc[:,1:2].values
y = datas.iloc[:,2].values
from sklearn.linear_model import LinearRegression
lin = LinearRegression()
lin.fit(X,y)
```

# Fitting a Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=4)
X_poly = poly.fit_transform(X)
poly.fit(X_poly,y)
lin2 = LinearRegression()
lin2.fit(X_poly,y)
```

# Visualizing the Linear Regression results using a scatter plot.

```python
plt.scatter(X,y,color='blue')
plt.plot(X, lin.predict(X), color='red')
plt.title('Linear Regression')
plt.xlabel('Temprature')
plt.ylabel('Pressure')
plt.show()
```

# Visualizing Polynomial Regression using a scatter plot.

```python
plt.scatter(X, y ,color='blue')
plt.plot(X, lin2.predict(poly.fit_transform(X)),
         color='red')
plt.title('Polynomial Regression')
plt.xlabel('Temprature')
plt.ylabel('Pressure')
plt.show()
```

# Predict new results with both Linear and Polynomial Regression

```python
pred = 110.0
predarray = np.array([[pred]])
lin.predict(predarray)
```

```python
pred2 = 110.0
pred2array = np.array([[pred2]])
lin2.predict(poly.fit_transform(pred2array))
```
